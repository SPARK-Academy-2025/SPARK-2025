# üêç *Python Classes & Objects*  

Python is a *powerful object-oriented programming* language.  

- üîπ In Python, *everything* is an object, possessing attributes (data) and methods (functions).  
- üèó A *class* serves as a *blueprint* for creating objects, defining their structure and behavior.  
- üîÑ Objects created from a class share common properties but hold unique data.

## 1.3Ô∏è‚É£ *The `__init__()` Function*  

The previous examples showed **basic classes and objects**, but real-world applications require **dynamic data handling**.  

To achieve this, we use the **built-in `__init__()` function**.  

### üîπ What is `__init__()`?
- üèó **A special method** that runs automatically when an object is created.  
- üîÑ **Used to assign values** to object properties.  
- ‚öôÔ∏è **Performs necessary setup** when initializing an object.  

### ‚ú® Example: Using `__init__()` in a `Patient` Class 


````python
class Patient:
    def __init__(self, name, age, test_result):
        self.name = name  # Assigning name to the object
        self.age = age  # Assigning age to the object
        self.test_result = test_result  # Assigning test result

# Now we can use the class named Patient to create objects:
# Create an object named patient1 with specific values
patient1 = Patient("John Doe", 45, "Positive")
patient2 = Patient("Peter", 56, "Negative")

# Accessing the attributes
print(f"Patient Name: {patient1.name}")
print(f"Patient Age: {patient1.age}")
print(f"Test Result: {patient1.test_result}")
print(f"Patient Name: {patient2.name}")
print(f"Patient Age: {patient2.age}")
print(f"Test Result: {patient2.test_result}")
````

## 1.4Ô∏è‚É£ *Object Methods*  

Objects can also contain **methods**.  
üîπ **Methods** are functions that belong to an object and define its behavior.  

### ‚ú® Example: Creating a Method in the `Patient` Class  

````python
class Patient:
    def __init__(self, name, age, test_result):
        self.name = name
        self.age = age
        self.test_result = test_result

    def display_info(self):
        return f"Patient Name: {self.name}, Age: {self.age}, Test Result: {self.test_result}"

# Create an object named patient1
patient1 = Patient("Alice", 30, "Negative")
patient2 = Patient("Peter", 56, "Negative")
# Call the method
print(patient1.display_info())
print(patient2.display_info())
````

## 1.5Ô∏è‚É£ *The `self` Parameter*  

The `self` parameter is a reference to the **current instance** of the class and is used to **access variables** that belong to the class.  

üí° **Key Points:**  
- `self` is always the **first parameter** of any function inside the class.  
- It does not have to be named `self`; you can use any name, but `self` is the convention. 

````python
class Patient:
    def __init__(custom, name, age):
        custom.name = name
        custom.age = age

    def display_info(custom):
        return f"Patient Name: {custom.name}, Age: {custom.age}"

# Create an object
patient1 = Patient("Alice", 30)

# Call the method
print(patient1.display_info())
````

````python
from datetime import datetime

class MedicalAppointment:
    def __init__(self, patient_name, doctor_name, appointment_date, reason):
        self.patient_name = patient_name
        self.doctor_name = doctor_name
        self.appointment_date = appointment_date
        self.reason = reason
    
    def schedule_appointment(self):
        """Schedules an appointment and confirms the details."""
        try:
            date_obj = datetime.strptime(self.appointment_date, "%Y-%m-%d %H:%M")
            confirmation_message = (
                f"Appointment Scheduled:\n"
                f"Patient: {self.patient_name}\n"
                f"Doctor: {self.doctor_name}\n"
                f"Date & Time: {date_obj.strftime('%A, %B %d, %Y at %I:%M %p')}\n"
                f"Reason: {self.reason}"
            )
            return confirmation_message
        except ValueError:
            return "Invalid date format. Please use 'YYYY-MM-DD HH:MM'."

# Object of MedicalAppointment Class
appointment = MedicalAppointment("John Doe", "Dr. Smith", "2025-03-10 14:30", "Routine Check-up")

print(appointment.schedule_appointment())
````

## 2Ô∏è‚É£ Python Inheritance  

**Inheritance** allows us to define a class that **inherits all methods and properties** from another class.  

### üîπ Key Concepts:  
- üèó **Parent Class** ‚Äì The class being inherited from (*also called the base class*).  
- üèó **Child Class** ‚Äì The class that inherits from another class (*also called the derived class*).  

Inheritance promotes **code reusability** and allows for **hierarchical class structures**. üöÄ  


## 2.1Ô∏è‚É£ *Creating a Parent Class*  

We can create a **Parent Class** that contains properties and methods which can be inherited by other classes.  

### ‚ú® Example: Defining a `Patient` Class  


````python
class Patient:
    def __init__(self, bmi, health_condition):
        self.bmi = bmi
        self.health_condition = health_condition

    def print_info(self):
        print(f"Patient BMI: {self.bmi}, Health Condition: {self.health_condition}")

# Create an object of the Patient class
patient1 = Patient(24.5, "Healthy")
patient1.print_info()
````

## 2.2Ô∏è‚É£ *Creating a Child Class*  

A **Child Class** inherits properties and methods from a **Parent Class**.  
To create a child class, pass the parent class as a parameter when defining it.  

### ‚ú® Example: Defining an `EmergencyPatient` Class 

````python
class EmergencyPatient(Patient):
    def __init__(self, bmi, health_condition, severity_level):
        super().__init__(bmi, health_condition)
        self.severity_level = severity_level

    def print_emergency_info(self):
        print(f"Patient BMI: {self.bmi}, Health Condition: {self.health_condition}, Severity Level: {self.severity_level}")

# Create an object of the EmergencyPatient class
patient2 = EmergencyPatient(30.2, "At Risk", "High")
patient2.print_emergency_info()
````

## 2.3 *Using the `super()` Function*  

Python provides the **`super()` function** to inherit all methods and properties from a **Parent Class** into a **Child Class**.  

### ‚ú® Example: Using `super()` in `EmergencyPatient` 

````python
# Parent class: MedicalScan
class MedicalScan:
    def __init__(self, patient_name, scan_type, date, diagnosis):
        self.patient_name = patient_name
        self.scan_type = scan_type
        self.date = date
        self.diagnosis = diagnosis

    def display_scan_info(self):
        return f"Patient: {self.patient_name}, Scan: {self.scan_type}, Date: {self.date}, Diagnosis: {self.diagnosis}"


# Child class 1: MRI Scan
class MRI(MedicalScan):
    def __init__(self, patient_name, date, diagnosis, magnetic_field_strength):
        super().__init__(patient_name, "MRI", date, diagnosis)
        self.magnetic_field_strength = magnetic_field_strength  # Measured in Tesla

    def display_mri_info(self):
        return f"{self.display_scan_info()}, Magnetic Field: {self.magnetic_field_strength}T"


# Child class 2: PET Scan
class PET(MedicalScan):
    def __init__(self, patient_name, date, diagnosis, radiotracer):
        super().__init__(patient_name, "PET", date, diagnosis)
        self.radiotracer = radiotracer  # Type of radiotracer used

    def display_pet_info(self):
        return f"{self.display_scan_info()}, Radiotracer: {self.radiotracer}"


# Child class 3: Ultrasound Scan
class Ultrasound(MedicalScan):
    def __init__(self, patient_name, date, diagnosis, frequency):
        super().__init__(patient_name, "Ultrasound", date, diagnosis)
        self.frequency = frequency  # Measured in MHz

    def display_ultrasound_info(self):
        return f"{self.display_scan_info()}, Frequency: {self.frequency} MHz"


# Creating objects
scan1 = MRI("Alice Johnson", "2025-02-15", "Brain Tumor", 3.0)
scan2 = PET("Michael Smith", "2025-02-16", "Lung Cancer", "Fluorodeoxyglucose (FDG)")
scan3 = Ultrasound("Emma Davis", "2025-02-17", "Kidney Stones", 5.0)

# Printing scan details
print(scan1.display_mri_info())
print(scan2.display_pet_info())
print(scan3.display_ultrasound_info())
````

## üîÑ Polymorphism in Object-Oriented Programming  

Polymorphism is a fundamental concept in **object-oriented programming (OOP)** that allows objects of different classes to be treated as objects of a common **parent class**.  

This enables us to write flexible and reusable code that can work with **different types of objects in a uniform way**.  

### ‚úÖ Why Polymorphism?  
- Allows **different classes** to share the **same method names** but have different implementations.  
- Promotes **code reusability** and **extensibility**.  
- Makes it easier to handle **different object types dynamically**.  

````python
# Parent class
class MedicalScan:
    def scan_type(self):
        return "General medical imaging scan"

# Child classes with specific implementations
class MRI(MedicalScan):
    def scan_type(self):
        return "Magnetic Resonance Imaging (MRI) - Uses magnetic fields"

class PET(MedicalScan):
    def scan_type(self):
        return "Positron Emission Tomography (PET) - Uses radiotracers"

class Ultrasound(MedicalScan):
    def scan_type(self):
        return "Ultrasound - Uses high-frequency sound waves"

# Function demonstrating polymorphism
def scan_details(scan):
    print(scan.scan_type())

# Creating objects of different scan types
mri_scan = MRI()
pet_scan = PET()
ultrasound_scan = Ultrasound()

# Calling the function with different scan types
scan_details(mri_scan)        # MRI-specific method
scan_details(pet_scan)        # PET-specific method
scan_details(ultrasound_scan) # Ultrasound-specific method
````

## üè• **OOP in Medical Data and Imaging** 

````python
import torch
from torch.utils.data import Dataset
import pandas as pd

class DiabetesDataset(Dataset):
    def __init__(self, csv_file):
        """
        Custom Dataset for loading diabetes data.
        Args:
        csv_file (str): Path to the CSV file containing the dataset.
        """
        self.data = pd.read_csv(csv_file)
        self.features = self.data.iloc[:, :-1].values  # All columns except the last one
        self.labels = self.data.iloc[:, -1].values     # Last column as labels

    def __len__(self):
        """Returns the total number of samples in the dataset."""
        return len(self.data)

    def __getitem__(self, idx):
        """
        Retrieves a single sample from the dataset.
        
        Args:
        idx (int): Index of the sample.

        Returns:
        tuple: (features, label) where features is a tensor of input data, and label is the corresponding target.
        """
        features = torch.tensor(self.features[idx], dtype=torch.float32)
        label = torch.tensor(self.labels[idx], dtype=torch.long)  # Assuming label is categorical (0 or 1)
        return features, label


dataset = DiabetesDataset("/kaggle/input/diabetes-data-set/diabetes.csv")
# print(f"Dataset size: {len(dataset)}")
# print(dataset.__len__())
sample_features, sample_label = dataset[0]
print(f"Sample Features: {sample_features}")
print(f"Sample Label: {sample_label}")
````

````python
# Custom Dataset for Brain Tumor Images
class BrainTumorDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        """
        Args:
            root_dir (str): Path to dataset directory (either Training or Testing).
            transform (callable, optional): Optional transforms to apply to images.
        """
        self.root_dir = root_dir
        self.transform = transform
        self.classes = ["glioma", "meningioma", "notumor", "pituitary"]
        
        # Collect all image paths and corresponding labels
        self.image_paths = []
        self.labels = []
        
        for label, class_name in enumerate(self.classes):
            class_dir = os.path.join(root_dir, class_name)
            for img_name in os.listdir(class_dir):
                img_path = os.path.join(class_dir, img_name)
                self.image_paths.append(img_path)
                self.labels.append(label)
    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert("RGB")  # Ensure 3-channel image
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label
````

````python
# Define transformations (resize, convert to tensor)
transform = transforms.Compose([
    transforms.Resize((125, 125)),  # Resize to match model input size
    transforms.ToTensor()
])

# Create Dataset Instances
train_dataset = BrainTumorDataset(root_dir="/kaggle/input/brain-tumor-mri-dataset/Training", transform=transform)
test_dataset = BrainTumorDataset(root_dir="/kaggle/input/brain-tumor-mri-dataset/Testing", transform=transform)

# Create DataLoaders
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=8, shuffle=False)

# Iterate through the train DataLoader
for images, labels in train_loader:
    print("Batch of images shape:", images.shape)
    print("Batch of labels:", labels)
    break  # Print only one batch
````

````python
import matplotlib.pyplot as plt

# Define label names
label_names = {0: "Glioma", 1: "Meningioma", 2: "No Tumor", 3: "Pituitary"}

# Get one batch of images and labels from the Train DataLoader
images, labels = next(iter(train_loader))

# Define the number of images to display
num_images = min(4, len(images))  # Display up to 8 images

# Create a figure
fig, axes = plt.subplots(1, num_images, figsize=(15, 5))

for i in range(num_images):
    image = images[i].permute(1, 2, 0)  # Change from (C 0, H 1, W 2) to (H, W, C)
    
    # Plot image
    ax = axes[i]
    ax.imshow(image)
    ax.axis("off")
    ax.set_title(f"Label: {label_names[labels[i].item()]}")
plt.show()
````
